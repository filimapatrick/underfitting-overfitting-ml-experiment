Import Modules

```python
import sys
import os

# Add parent directory to path (where the Python modules are located)
sys.path.append(os.path.abspath(".."))

from data_generation import generate_nonlinear_data, split_data
from model_training import train_polynomial_regression, evaluate_model
from visualization import plot_regression_curve, plot_error_curve
```

Generate Data

```python
X, y = generate_nonlinear_data()
X_train, X_test, y_train, y_test = split_data(X, y)
```

Train 3 Models

```python
degrees = [1, 5, 15]

train_errors = []
test_errors = []

for degree in degrees:
    
    model, poly = train_polynomial_regression(X_train, y_train, degree)

    train_mse, _ = evaluate_model(model, poly, X_train, y_train)
    test_mse, _ = evaluate_model(model, poly, X_test, y_test)

    train_errors.append(train_mse)
    test_errors.append(test_mse)

    print(f"Degree {degree}")
    print(f"Training MSE: {train_mse:.4f}")
    print(f"Testing MSE: {test_mse:.4f}")
    print("-" * 30)

    plot_regression_curve(
        X, y, model, poly, degree,
        save_path=f"../results/plots/degree_{degree}.png"
    )
```

Plot Error Curve

```python
plot_error_curve(
    degrees,
    train_errors,
    test_errors,
    save_path="../results/plots/error_curve.png"
)
```
